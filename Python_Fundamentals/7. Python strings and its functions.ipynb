# Strings

Text data in Python is referred to as strings. It is formed by enclosing characters in single quotes(''), double quotes(" ") and for multiline strings using triple quotes("""  """).

```python
s = "I love coding"
print(s)
```

### String Iteration

```python
for ch in s:
    print(ch)
```

```python
index = 0
while index<len(s):
    print(s[index])
    index = index +1
```
