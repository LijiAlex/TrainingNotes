### Loop
A Loop is used to execute a group of instructions or a block of code multiple times, without writing it repeatedly. 



Print numbers from 1 - 5.

```python
print(1)
print(2)
print(3)
print(4)
print(5)
```

```python
num = 1
while num<=10:
    print(num)
    num = num + 1
print("printed all numbers")
```

Print numbers from 10-1

```python
num = 5
while num>=1:
    print(num)
    num -= 1
print("printed all numbers")
```

### Structure of while loop
1. initialization

Multiplication table of a number

```python
num = int(input("Enter a number: "))
i = 1
while i<= 10:
    print(f"{num} X {i} = {num*i}")
    i = i+1
```

### Infinite loop

```python
while True:
    print(1)
```

### break a loop

Get numbers from user till 0 is entered

```python
while True:
    num = int(input("Enter a number: "))
    if num==0:
        break
    else:
        print(num)
```

### continue

print numbers from 1-20 except multiples of 3

```python
num = 1
while num<=10:
    if num%3==0:
        num += 1
        continue
    print(num)
    num += 1
```
