### Loop
A Loop is used to execute a group of instructions or a block of code multiple times, without writing it repeatedly.

Print numbers from 1 - 5.

```python
print(1)
print(2)
print(3)
print(4)
print(5)
```

```python
num = 1  # initialization
while num<=1000:  # condition checking
    print(num)    # action
    num = num + 1   #updation
print("printed all numbers")
```

Print numbers from 10-1

```python
num = 5  # initialization
while num>=1: # condition checking
    print(num)
    num-=1  # updation
print("printed all numbers")
```

### Structure of while loop
1. initialization
2. condition checking
3. updation

Multiplication table of a number

```python
num = int(input("Enter a number: "))
i = 1 
while i<= 10000:
    print(f"{num} X {i} = {num*i}")
    i = i+1
```

### Infinite loop

```python
while True: #infinite loop
    print(1)
```

### break a loop

Get numbers from user till 0 is entered

```python
while True: # infinite loop
    num = int(input("Enter a number: "))
    if num==0: # condition
        break # break from loop if condition is true
    else: # if condition is false
        print(num)
```

### continue

print numbers from 1-20 except multiples of 3

```python
1 2 4 5 7 8 10
```

```python
num = 1 # initialization
while num<=20: #loop condition
    if num%3==0:  # another check
        num += 1
        continue  # continue to beginning of loop after skipping rest of teh sentences
    print(num)
    num += 1
```

Q2. The apartment numbers in a dwelling is A1, A2, A3...A20. Print these apartment no's in a single line.

```python
text_part = 'A'
number_part = 1
while number_part<=20:  
  print(text_part+ str(number_part), end=" ")
  number_part +=1  
```

Q5. The apartment numbers in a dwelling is A1, A2, A3...A20, B1 ..B20, ...C1..C20...D1..D20..E1..E5. Print these apartment nos with each series in a line.

```python
text_part = 'A'
number_part = 1
while text_part<='E' and number_part<=20:  
  print(text_part+ str(number_part), end=" ")
  number_part +=1
  if number_part == 21:
    number_part = 1
    text_part = chr(ord(text_part) + 1)
    print()
  if text_part == 'E' and number_part > 5:
    break
```

Q6. The apartment numbers in a dwelling is A1, A2, A3...A20, B1 ..B20, ...C1..C20...D1..D20..E1..E5. But usually 13 is considered to be unlucky. So skip this number. ie A13, B13, C13 and D13 won't be there. Print these apartment nos with each series in a line.

```python
text_part = 'A'
number_part = 1
while text_part<='E' and number_part<=20:  
  if number_part == 13:
    number_part +=1
    continue
  print(text_part+ str(number_part), end=" ")
  number_part +=1
  if number_part == 21:
    number_part = 1
    text_part = chr(ord(text_part) + 1)
    print()
  if text_part == 'E' and number_part > 5:
    break
```
