What is the difference between the below statements
<br>Statement 1: "Meena got 95% in Maths"
<br>Statement 2: "Meena got 95 percentile in Maths"?

1. Statement 1 means Meena got 95 marks out of 100.
2. Statement 2 means Meena scored more than 95% students who appeared for the exam. It doesn't indicate Meena's marks over here.

The key difference between percentage and percentile is that 
* percentage is a mathematical value presented out of 100 and percentile is the percent of values below a specific value.
* percentage shows quantity where as percentile shows rank.
* percentage is standalone where as percentile is in comparison with others.

### Quatriles

Quartiles divide the data into four groups, each containing an equal number of values. Quartiles are divided by the 25th, 50th, and 75th percentile, also called the first, second and third quartile.


### 5 number Summary

The five number summary contains: minimum, maximum, median, 25th percentile value and 75th percentile value.

```python
import pandas as pd
df = pd.DataFrame(data_set)
df.describe()
```

```python
def summary(data):
    minVal = np.min(data)
    Q1, median, Q3 = np.percentile(data, [25, 50, 75])
    maxVal = np.max(data)
    return (minVal, Q1, median, Q3, maxVal)
```

```python
minval, Q1, median, Q3, maxval = summary(data_set)
minval, Q1, median, Q3, maxval
```

```python
IQR = Q3 - Q1
IQR
```
