### What is statistics?

Science of collecting, organizing and analyzing data to derive some meaningful conclusion.

### Why statistics?
1. Derive some meaningful information from raw data.
2. Comparison tool.
3. Can quantify the uncertainity.
4. Recognize patterns.

### Population data vs sample data
* A population is the entire group that you want to draw conclusions about.
* A sample is the specific group that you will collect data from.

### Two types

Statistics have majorly categorised into two types: 
* Descriptive statistics- Meaningful Summary of data from  a given dataset.
<br> Eg:- Avg weight, Maximum score etc 
* Inferential statistics- Allows us to use information collected from a sample to make decisions, predictions or inferences from a population. It grants us permission to give statements that goes beyond the available data or information.
<br> Eg:- Exit poll, Survey results



# Descriptive Statistics
Descriptive statistics are used to describe the basic features of the data in a study. They provide simple summaries about the sample and the measures.


## Measures of central tendency

* A measure of central tendency is a value that attempts to describe a set of data by identifying the central position within that set of data.
* There are three main measures of central tendency: 
1. the mode, 
2. the median 
3. and the mean.


### Mean
The mean is the mathematical average of a set of two or more numbers.


```python
import numpy as np
data_set = [6, 4, 10, 3, 7]
np.mean(data_set)
```

### Median
Median represents the middle value for any group. It is the point at which half the data is more and half the data is less.


```python
data_set = [1, 3, 3, 6, 7, 8, 9]
np.median(data_set)
```

```python
data_set = [1, 2, 3, 4, 5, 6, 8, 9]
np.median(data_set)
```

### Mode

The data point having the maximum frequency or the data point that is repeted most.



```python
import statistics
data_set = [1, 3, 3, 6, 7, 8, 9]
statistics.mode(data_set)
```

## Measures of spread

Describes how much the data is spread.
1. Range
2. Standard Deviation
3. Variance
4. Inter quartile range
5. Z Score

### Range
The range is the difference between the highest and lowest values in a set of numbers.


### Standard Deviation & Variance
* The standard deviation is a statistic that measures the dispersion of a dataset relative to its mean.
* SD is calculated as the square root of the variance. 
* The more spread out the data, the higher the standard deviation.



```python
data_set = [1, 4, 5, 6, 2, 3, 5, 6, 7, 8]
np.std(data_set)
```

```python
np.var(data_set)
```

### Inter Quartile Range

* Dispertion with respect to median.
* Spread of middle 50% data.




### Z Score

Z-score is measured in terms of standard deviations from the mean.
