### Variable Types

* Categorical - Names/labels that cannot be quantified. Eg:- Marital Status, Gender, Eye Colour etc
* Numeric- Numeric values which can be quantified. Eg:- Population, Team size, height weight


# Types of distributions

### Gaussian distribution/normal distribution

Gaussian distribution (also known as normal distribution) is a bell-shaped curve, with an equal number of measurements above and below the mean value.


### Standard normal distribution

The standard normal distribution, also called the z-distribution, is a special normal distribution where the mean is 0 and the standard deviation is 1. Any normal distribution can be standardized by converting its values into z-scores. Z-scores tell you how many standard deviations from the mean each value lies.



### Log normal distribution
A log-normal distribution is a continuous probability distribution of a random variable whose logarithm is normally distributed.
