* It is used to convert a normal distribution to standard normal distribution.

### Z table

The area under the whole of a normal distribution curve is 1, or 100 percent. The z-table helps by telling us what percentage is under the curve at any particular point.
<br>Reference: https://www.ztable.net/
<br>Positive Z-table : To know the area to the left of a positive Z-Score.

<br>Negative Z-table: To know the area to the left of a negative Z-Score.


### What is the probability of anyone scoring less than 40? mean = 30, standard deviation =4

<br>Value corresponding to 2.5(z score of 40) from positive Z table is 0.99375.
<br>So, 99.38% scored less than 40.

### What is the probability of anyone scoring more than 40? mean = 30, standard deviation =4


<br>Value corresponding to 2.5(z score of 40) from left Z table is 0.99375.
<br>Area to the right is 1-0.9938 = .00625
<br>So, 0.63% scored more than 40.

### What is the probability of anyone scoring between 30 and 40?mean = 30, standard deviation =4



<br> Value corresponding to 2.5(z score of 40) from positive Z table is 0.99375.
<br> Value corresponding to 0(z score of 30)from positive Z table is 0.5.
<br> Subtracting we get 0.99375 - 0.5 = 0.49375
<br>So, 49.38% scored between 30 and 40.

### Steps to look an Z table

1. Find Z Score of the point.
2. To find the area to the left of the point(positive), look up corresponding value in the positive Z table.

3. To find the area to the right of the point(positive),look up corresponding value in the positive Z table and subtract it from 1 or find the corresponding value in the negative Z table.

4. To find the area to the left of the point(negative), look up corresponding value in the negative Z table .

5. To find the area to the right of the point(negative), look up corresponding value in the negative Z table and subtract it from 1 or find the corresponding value in the positive Z table.

6. To find the area between two positive scores, look up the areas that go with the two z scores from positive z table. Next subtract the smaller area from the larger area.

7. To find the area between two negative scores, look up the areas that go with the two z scores from negative z table. Next subtract the smaller area from the larger area.

### Python code to get value from z table

Function: stats.norm

* loc refers to mean
* scale refers to standard deviation

```python
stats.norm(loc = 50000 , scale = 20000).cdf(40000)  # Cumulative distribution function
```

### What is the probability of anyone scoring less than 40? mean = 30, standard deviation =4


```python
stats.norm(loc = 30 , scale = 4).cdf(40)
```

### What is the probability of anyone scoring more than 40? mean = 30, standard deviation =4


```python
1 - stats.norm(loc = 30 , scale = 4).cdf(40)
```

```python
stats.norm(loc = 30 , scale = 4).sf(40) # 1 -cdf, Survival function
```

### What is the probability of anyone scoring between 30 and 40? mean = 30, standard deviation =4



```python
area = stats.norm(loc = 30 , scale = 4).cdf([30, 40])
area
```
